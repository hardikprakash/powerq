```python
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(df_trunc.drop(['Datetime'], axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```
